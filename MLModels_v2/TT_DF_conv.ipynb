```python
import csv
import pandas as pd

def parse_input_to_csv(input_string, output_file):
    # Define the headers
    headers = ["Sl.No", "Class Group", "Course", "L T P J C", "Category", 
               "Course Option", "Class Id", "Slot/Venue", "Faculty Details", 
               "Registered / Updated Date & Time", "Attendance Date/Type", 
               "Status & Ref. No."]
    
    # Split the input string into lines and clean up extra spaces/tabs
    lines = [line.strip() for line in input_string.split('\n') if line.strip()]
    
    # Initialize a list to hold the parsed rows
    parsed_rows = []
    
    i = 0
    while i < len(lines):
        if lines[i].isdigit():  # Sl.No
            row = [lines[i]]  # Sl.No
            i += 1
            row.append(lines[i])  # Class Group
            i += 1
            row.append(lines[i])  # Course
            i += 1
            i += 1  # Skipping extra line (e.g., (Theory Only), (Lab Only))
            row.append(lines[i])  # L T P J C
            i += 1
            row.append(lines[i])  # Category
            i += 1
            row.append(lines[i])  # Course Option
            i += 1
            row.append(lines[i])  # Class Id
            i += 1
            row.append(lines[i])  # Slot/Venue
            i += 1
            row.append(lines[i])  # Faculty Details
            i += 1
            row.append(lines[i])  # Registered/Updated Date & Time
            i += 1
            row.append(lines[i])  # Attendance Date/Type
            i += 1
            row.append(lines[i])  # Status & Ref. No.
            i += 1
            parsed_rows.append(row)
        else:
            i += 1

    # Write to the CSV file
    with open(output_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(headers)  # Writing the headers
        writer.writerows(parsed_rows)  # Writing the parsed rows


def csv_to_dataframe(output_file):
    # Load CSV into a pandas DataFrame
    df = pd.read_csv(output_file)
    
    # Retrieve only "Course" and "Slot/Venue" columns
    df_filtered = df[["Course", "Slot/Venue"]]
    
    return df_filtered

# Example usage
input_string ="""Sl.No	Class Group	Course	L T P J C	Category	Course Option	Class Id	Slot/ Venue	Faculty Details	Registered / Updated Date & Time	Attendance Date/ Type	Status & Ref. No.
1

General Freshers

BCHY101L - Engineering Chemistry

( Theory Only )

3 0 0 0 3.0

Foundation Core	
Regular

VL2024250106782

G2+TG2 -


PRP122

MADHVESH PATHAK -


SAS

27-Jul-2024 23:39

28-Jul-2024
- Manual

Registered and Approved
2

General Freshers

BCHY101P - Engineering Chemistry Lab

( Lab Only )

0 0 2 0 1.0

Foundation Core	
Regular

VL2024250107517

L7+L8 -


SMVG027

NAMRATA DEKA -


SAS

27-Jul-2024 23:39

28-Jul-2024
- Manual

Registered and Approved
3

General Freshers

BCSE101E - Computer Programming: Python

( Embedded Theory )

1 0 0 0 1.0

Foundation Core	
Regular

VL2024250107822

TCC2 -


PRP122

RAJARAJAN G -


SCOPE

27-Jul-2024 23:39

28-Jul-2024
- Manual

Registered and Approved
4

General Freshers

BCSE101E - Computer Programming: Python

( Embedded Lab )

0 0 4 0 2.0

Foundation Core	
Regular

VL2024250107979

L11+L12+L25+L26 -


PRP117a

RAJARAJAN G -


SCOPE

27-Jul-2024 23:39

28-Jul-2024
- Manual

Registered and Approved
5

General Freshers

BCSE101N - Introduction to Engineering

( Project )

0 0 0 0 1.0

Non-graded Core Requirement	
Regular

VL2024250108301

NIL -


NIL

ACADEMICS -


ACAD

27-Jul-2024 23:39

28-Jul-2024
- Manual

Registered and Approved
6

General Freshers

BEEE102L - Basic Electrical and Electronics Engineering

( Theory Only )

3 0 0 0 3.0

Foundation Core	
Regular

VL2024250106439

C2+TC2 -


PRP122

SUPRAVA CHAKRABORTY -


TIFAC

27-Jul-2024 23:39

28-Jul-2024
- Manual

Registered and Approved
7

General Freshers

BEEE102P - Basic Electrical and Electronics Engineering Lab

( Lab Only )

0 0 2 0 1.0

Foundation Core	
Regular

VL2024250107008

L27+L28 -


TT344

SUPRAVA CHAKRABORTY -


TIFAC

27-Jul-2024 23:39

28-Jul-2024
- Manual

Registered and Approved
8

General Freshers

BMAT101L - Calculus

( Theory Only )

3 0 0 0 3.0

Foundation Core	
Regular

VL2024250106327

E2+TE2 -


PRP122

ARUN CHOUDHARY -


SAS

27-Jul-2024 23:39

28-Jul-2024
- Manual

Registered and Approved
9

General Freshers

BMAT101P - Calculus Lab

( Lab Only )

0 0 2 0 1.0

Foundation Core	
Regular

VL2024250106721

L15+L16 -


PRP118

ARUN CHOUDHARY -


SAS

27-Jul-2024 23:39

28-Jul-2024
- Manual

Registered and Approved
10

General Freshers

BSTS201P - Qualitative Skills Practice I

( Soft Skill )

0 0 3 0 1.5

Foundation Core	
Regular

VL2024250106203

F2+TF2 -


PRP309

FACE (APT) -


SSL

27-Jul-2024 23:39

28-Jul-2024
- Manual

Registered and Approved"""
output_file = "courses.csv"

# Step 1: Parse input to CSV
parse_input_to_csv(input_string, output_file)

# Step 2: Convert CSV to DataFrame and filter columns
df_filtered = csv_to_dataframe(output_file)

# Output the filtered DataFrame
print(df_filtered)
```

```python
import pandas as pd

def process_dataframe(df, timetable_dict):
    # Remove hyphens and spaces from Slot/Venue column
    df['Slot/Venue'] = df['Slot/Venue'].str.replace('-', '').str.replace(' ', '')
    
    # Split Slot/Venue into a list
    df['Slot/Venue'] = df['Slot/Venue'].str.split('+')
    
    # Create a new Day-time column
    def get_day_time(slots):
        day_time = []
        for slot in slots:
            if slot in timetable_dict:
                day_time.extend(timetable_dict[slot])
        return day_time
    
    df['Day-time'] = df['Slot/Venue'].apply(get_day_time)
    
    return df

# Assuming your dataframe is called 'df' and your dictionary is 'timetable_dict'
# df = pd.read_csv('your_csv_file.csv')  # Uncomment this line if you're reading from a CSV

# Process the dataframe
result_df = process_dataframe(df_filtered, timetable_dict)

# Display the result
print(result_df[['Course', 'Slot/Venue', 'Day-time']])
```
